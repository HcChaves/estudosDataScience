# Load da base para estudo

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
print(df.head())
```

```python
# verificandpo tamanho do df
print(f'Número de linhas: {df.shape[0]}\nNúmero de variáveis: {df.shape[1]}')
```

```python
# sumarizando o df
df.describe()
```

```python
# verificando as distribuições das variaveis
df.hist()
plt.show()
```

```python
# Gráfico de dispersão com linha de regressão
sns.regplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
# Gráfico de dispersão com linha de regressão + male
sns.scatterplot(data=df,x='total_bill',y='tip',hue='time', style='day')
plt.show()
```

```python
# verificando tipos das variáveis
df.info()
```

```python
# selecionando somente variaveis numericas
df_numerics = df.select_dtypes(include=['float64','int64'])
```

```python
# calculando a correlação
corr = df_numerics.corr()

# plotando o heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1).xaxis.tick_top()
```

```python
sns.barplot(data=df)
plt.show()
```
